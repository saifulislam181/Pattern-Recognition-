```python
import random
import numpy as np
from google.colab import drive
from numpy import genfromtxt
from scipy.spatial import distance
from collections import Counter
drive.mount("/content/gdrive")
```

```python
data_path = '/content/gdrive/MyDrive/Colab Notebooks/iris.csv'
my_data = genfromtxt(data_path, delimiter=',')
```

```python
k = 10
```

```python
def feature_output(samples):
  sx = []
  sy = []

  for i in samples:
    templist = []
    templist.extend(i[0:4])
    sx.append(templist.copy())
    sy.append(i[-1])
  return sx, sy
```

```python
Train_set = []
Val_set = []
Test_set = []

random.shuffle(my_data)

for s in my_data:
  temp = random.uniform(0, 1)
  if (0 <= temp and temp <= 0.7):
    Train_set.append(s)
  elif (0.7 < temp and temp <= 0.85):
    Val_set.append(s)
  else:
    Test_set.append(s)
```

```python

vin, vout = feature_output(Val_set)
tin, tout = feature_output(Train_set)

vin = np.array(vin)
tin = np.array(tin)
correct = 0

for v in range(0, len(Val_set)):
  L = []
  nearest = []
  for t in range(0, len(Train_set)):
    d = np.sqrt(np.sum(np.square(vin[v] - tin[t])))
    L.append([d, t])
  L.sort()
  ksamples = L[0: k]
  for d, v in ksamples:
    nearest.append(tout[v])
  mk = Counter(nearest).most_common()[0][0]
  ms = Counter(tout).most_common()[0][0]
  if mk==ms:
    correct += 1

acc = correct / len(Val_set) * 100
print(acc)
```
